```python
import csv
from datetime import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
stroop_data = pd.read_csv('stroopdata.csv')
Congruent_count=stroop_data['Congruent'].count()
Congruent_mean=stroop_data['Congruent'].mean()
Congruent_median=stroop_data['Congruent'].median()
Congruent_max=stroop_data['Congruent'].max()
Congruent_min=stroop_data['Congruent'].min()
print(Congruent_count)
print(Congruent_mean)
print(Congruent_median)
print(Congruent_max)
print(Congruent_min)
```

```python
plt.figure(figsize=(8,6))
value_range=Congruent_max-Congruent_min
n_bins=5
bin_width = float(value_range) / n_bins
bins = [i*bin_width + Congruent_min for i in range(n_bins+1)]

plt.hist(stroop_data['Congruent'],bins=bins)
plt.xlabel('time')
plt.ylabel('count')
plt.title('number of Congruent')
```

```python
Incongruent_count=stroop_data['Incongruent'].count()
Incongruent_mean=stroop_data['Incongruent'].mean()
Incongruent_median=stroop_data['Incongruent'].median()
Incongruent_max=stroop_data['Incongruent'].max()
Incongruent_min=stroop_data['Incongruent'].min()
print(Incongruent_count)
print(Incongruent_mean)
print(Incongruent_median)
print(Incongruent_max)
print(Incongruent_min)
```

```python
plt.figure(figsize=(8,6))
value_range=Incongruent_max-Incongruent_min
n_bins=5
bin_width = float(value_range) / n_bins
bins = [i*bin_width + Congruent_min for i in range(n_bins+1)]

plt.hist(stroop_data['Incongruent'],bins=bins)
plt.xlabel('time')
plt.ylabel('count')
plt.title('number of Incongruent')
```
